```python
import pandas as pd

gt = pd.read_csv('ground_truth.csv')

m1 = pd.read_csv('predicted_value_1.csv')
m2 = pd.read_csv('predicted_value_2.csv')
m3 = pd.read_csv('predicted_value_3.csv')

```

```python
def checkDuplicateID(df):
    l = list(df['order_id'])
    return len(l) == len(set(l))

print(checkDuplicateID(gt))
print(checkDuplicateID(m1))
print(checkDuplicateID(m2))
print(checkDuplicateID(m3))
```

```python
# check if the order id is different

l0 = gt['order_id']
l1 = m1['order_id']
l2 = m2['order_id']
l3 = m3['order_id']

print((l0==l1).all())
print((l0==l2).all())
print((l0==l3).all())
```

```python
# visualize data
ground = gt.plot(kind = 'scatter', x='order_id', y ='final_price', title = 'ground truth')
model_1 = m1.plot(kind = 'scatter', x='order_id', y ='final_price', title = 'model 1')
model_2 = m2.plot(kind = 'scatter', x='order_id', y ='final_price', title = 'model 2')
model_3 = m3.plot(kind = 'scatter', x='order_id', y ='final_price', title = 'model 3')
fig1 = ground.get_figure()
fig2 = model_1.get_figure()
fig3 = model_2.get_figure()
fig4 = model_3.get_figure()
fig1.savefig('ground_truth.png')
fig2.savefig('model1.png')
fig3.savefig('model2.png')
fig4.savefig('model3.png')
```

```python
import numpy as np
a = np.array(gt['final_price'])
dis_gt = np.var(gt['final_price'])
mean_gt = np.mean(gt['final_price'])
new_x = (a-mean_gt)/dis_gt
print(np.max(new_x))
print(dis_gt,mean_gt)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
# x = new_x
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
ground = gt['final_price'].plot.kde()
fig1 = ground.get_figure()
model1 = m1['final_price'].plot.kde()
fig2 = model1.get_figure()
model2 = m2['final_price'].plot.kde()
fig3 = model2.get_figure()
model3 = m3['final_price'].plot.kde()
fig4 = model3.get_figure()

fig1.savefig('ground_truth_pdf.png')
fig2.savefig('model1_pdf.png')
fig3.savefig('model2_pdf.png')
fig4.savefig('model3_pdf.png')
m1['final_price'].plot.kde()
m2['final_price'].plot.kde()
m3['final_price'].plot.kde()
```
